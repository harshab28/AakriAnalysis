# **Model Build**

```python
ds = pd.read_excel("Analyst case study dataset 1.xls")

df = ds.drop(columns=["FirstName", "Email", "VendorLeadID", "IP Address", "ReferralDomain", "SearchQuery", "ReferralURL", "ReferralURL Parameters", "LandingPageURL", "Landing Page URL Parameters"], )

df['LeadCreated'] = df['LeadCreated'].dt.month.astype(str).str.zfill(2)
```

```python
status_mapping = {
    'Closed': 3,
    'EP Sent': 2,
    'EP Received': 2,
    'EP Confirmed': 2,
    'Contacted - Invalid Profile': 1,
    'Contacted - Doesn\'t Qualify': 1,
    'Unable to contact - Bad Contact Information': 0,
}
df['CallStatus'] = df['CallStatus'].map(status_mapping)
df['CallStatus'] = df['CallStatus'].fillna(0)
df['AddressScore'] = df['AddressScore'].fillna(0)
df['PhoneScore'] = df['PhoneScore'].fillna(0)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
for i in df.columns:
  df[i] = encoder.fit_transform(df[i])

df.fillna(0)
```

```python
y = df.iloc[:, 1].values

df.drop('CallStatus', axis=1, inplace=True)

X = df.iloc[:, :].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=2)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
gnb = GaussianNB()
cv = cross_val_score(gnb,X_train,y_train,cv=5)
print(cv)
print(cv.mean())
```

```python
lr = LogisticRegression(max_iter = 2000)
cv = cross_val_score(lr,X_train,y_train,cv=5)
print(cv)
print(cv.mean())
```

```python
dt = tree.DecisionTreeClassifier(random_state = 1)
cv = cross_val_score(dt,X_train,y_train,cv=5)
print(cv)
print(cv.mean())
```

```python
knn = KNeighborsClassifier()
cv = cross_val_score(knn,X_train,y_train,cv=5)
print(cv)
print(cv.mean())
```

```python
rf = RandomForestClassifier(random_state = 1)
cv = cross_val_score(rf,X_train,y_train,cv=5)
print(cv)
print(cv.mean())
```

```python
svc = SVC(probability = True)
cv = cross_val_score(svc,X_train,y_train,cv=5)
print(cv)
print(cv.mean())
```

```python
from sklearn.ensemble import VotingClassifier
voting_clf = VotingClassifier(estimators = [('lr',lr),('knn',knn),('rf',rf),('gnb',gnb),('svc',svc)], voting = 'soft')
```

```python
cv = cross_val_score(voting_clf,X_train,y_train,cv=5)
print(cv)
print(cv.mean())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# RandomForest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions and evaluation
y_pred_rf = rf_model.predict(X_test)

print(f'Accuracy: {accuracy_score(y_test, y_pred_rf)}')
print(f'Classification Report:\n{classification_report(y_test, y_pred_rf)}')

# Feature importance
importances = rf_model.feature_importances_
features = df.columns

feature_importance = pd.DataFrame({'Feature': features, 'Importance': importances})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)
print(feature_importance)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

# Gradient Boosting model
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train, y_train)

# Predictions and evaluation
y_pred_gb = gb_model.predict(X_test)

print(f'Accuracy: {accuracy_score(y_test, y_pred_gb)}')
print(f'Classification Report:\n{classification_report(y_test, y_pred_gb)}')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Select numerical columns for clustering
df_num = df[['AddressScore', 'PhoneScore']]  # Add more features as needed

# Standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_num)

# Determine the optimal number of clusters using the Elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Fit KMeans with optimal clusters
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=42)
df['Cluster'] = kmeans.fit_predict(df_scaled)

# Analyze clusters
df.groupby('Cluster').mean()  # You can check the characteristics of each cluster to identify improvement areas
```

# **Statiscal Significance**

```python
ds = pd.read_excel("Analyst case study dataset 1.xls")

df = ds.drop(columns=["FirstName", "Email", "VendorLeadID", "IP Address", "ReferralDomain", "SearchQuery", "ReferralURL", "ReferralURL Parameters", "LandingPageURL", "Landing Page URL Parameters"], )

df['LeadCreated'] = df['LeadCreated'].dt.month.astype(str).str.zfill(2)
```

```python
status_mapping = {
    'Closed': 3,
    'EP Sent': 2,
    'EP Received': 2,
    'EP Confirmed': 2,
    'Contacted - Invalid Profile': 1,
    'Contacted - Doesn\'t Qualify': 1,
    'Unable to contact - Bad Contact Information': 0,
}
df['CallStatus'] = df['CallStatus'].map(status_mapping)
df['CallStatus'] = df['CallStatus'].fillna(0)
df['AddressScore'] = df['AddressScore'].fillna(0)
df['PhoneScore'] = df['PhoneScore'].fillna(0)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
for i in df.columns:
  df[i] = encoder.fit_transform(df[i])

df.fillna(0)
```

```python
from scipy.stats import ttest_ind

# Splitting dataset into two periods
df_early = df[df['WidgetName'] < 6]  # Adjust date as needed
df_late = df[df['WidgetName'] >= 6]

# Compare mean call success counts
t_stat, p_value = ttest_ind(df_early['CallStatus'].dropna(), df_late['CallStatus'].dropna(), equal_var=False)

print(f'T-Test Statistic: {t_stat}')
print(f'P-Value: {p_value}')

# Interpretation
if p_value < 0.05:
    print("Statistically significant difference in lead quality over time.")
else:
    print("No significant difference in lead quality over time.")
```

```python
import scipy.stats as stats

# Select two categorical variables (e.g., 'State' and 'CallStatus')
contingency_table = pd.crosstab(df['WidgetName'], df['CallStatus'])

# Perform Chi-Square Test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print(f'Chi-Square Statistic: {chi2}')
print(f'P-value: {p}')  # If p < 0.05, reject null hypothesis (significant relationship)
```

```python
from scipy.stats import ttest_ind

# Select two groups (e.g., Leads from 'Google' vs 'Yahoo')
group1 = df[df['Partner'] == 'Google']['CallStatus']
group2 = df[df['Partner'] == 'Yahoo']['CallStatus']

# Perform T-Test
t_stat, p_value = ttest_ind(group1, group2, nan_policy='omit')

print(f'T-Statistic: {t_stat}')
print(f'P-value: {p_value}')  # If p < 0.05, reject null hypothesis (significant difference)
```

```python
from scipy.stats import ttest_rel

# Define two time periods for comparison
before = df[df['LeadCreated'] <= 6]['CallStatus']  # Before a threshold date
after = df[df['LeadCreated'] > 6]['CallStatus']    # After a threshold date
print(len(before), len(after))
# Perform Paired T-Test
t_stat, p_value = ttest_rel(df_early, df_late, nan_policy='omit')

print(f'Paired T-Statistic: {t_stat}')
print(f'P-value: {p_value}')  # If p < 0.05, reject null hypothesis (lead quality has changed)
```
